6. Filter the DataFrame on **one specific module** (your choice) and plot its **attendance rate over time**
   (x = Date, y = the module's average attendance).



````python
# your code here
mfilter = a_data[a_data['Module Name'] == "Project Management"]
#i need to first get average attendance for each day, whats important here is date and the attended column 
#as well as the number of people  in the class
#this can easily be done by summing the atended column for a given date and putting it against the total number of people in the class
#so first i need to find the number of people in the class
classsize = len(mfilter['Person Code'].unique())
````

7. **[For L6 students only]** Add a column, titled "Student Overall Attendance", with the average attendance for the student. Add a column, titled "Standardised Student Overall Attendance", with the z-scores of "Student Overall Attendance".

````python
#attendance rate for each student
student_attendance = a_data.groupby('Person Code').agg({'Attended':'sum','NotAttended':'sum'}).reset_index()
student_attendance['Student overall attendance'] = (student_attendance['Attended'] / (student_attendance['Attended'] + student_attendance['NotAttended']))

#z scores
mean_attendance = student_attendance['Student overall attendance'].mean()
std_attendance = student_attendance['Student overall attendance'].std()
student_attendance['Standardised student overall attendance'] = ((student_attendance['Student overall attendance'] - mean_attendance) / std_attendance)

#add to original dataframe
a_data = a_data.merge(student_attendance[['Person Code', 'Student overall attendance', 'Standardised student overall attendance']],on='Person Code',how='left')
a_data.tail()
````

### **Exercise 4 â€” Comparing Modules**



1. Compute the **mean** and **standard deviation** of attendance per module using pandas `groupby`. Consult the documentation if needed.


````python
# your code here
import scipy.stats as stats
grouped = a_data.groupby(['Module Code','Module Name'])

#empty list for results
results = []

#loop through each group
for (module_code, module_name), group in grouped:
    mean_attendance = group['Attended'].mean()
    std = group['Attended'].std()
    n = len(group)
    results.append({'Module Code': module_code,'Module Name': module_name,'Mean Attendance': mean_attendance,'Standard Deviation': std,'Class Size': n,})

#convert results to df
attendance_stats = pd.DataFrame(results)
attendance_stats.head()
````

2. Create a `seaborn.barplot` of mean attendance per module with 95% confidence intervals.


````python
# your code here
#because of the way i did the previous question and because this question requires seaborn, i now have to return to previous data :(
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(data=a_data, x='Module Name', y='Attended', errorbar=('ci',95))

plt.title('Mean Attendance per Module, 95% CI')
plt.xticks(rotation=90)
plt.show()
````

### **Exercise 5 â€” Attendance Relationships**



1. Answer the question: **Do students with low overall attendance tend to attend modules that have low attendance?**.

2. **[For L5 students only]**. Choose between a correlation analysis or a linear regression analysis to answer the question.

3. **[For L6 students only]** Run both correlation analysis and linear regression analysis to answer the same question. Do they tell the same story?

> Tips:
> 1. Compute the mean attendance rate of each module. Can you re-use some of the `groupby` code you wrote above?
> 2. For each module, compute the average overall attendance of the students who attended that module.
> 3. Correlation and linear regression can both help you answer the question. 
> 4. Report main statistic and p-value(s).


````python
# your code here
#for this i need to compare a student's attendance in a module vs the module average.
newdf = a_data.merge(attendance_stats[['Module Code', 'Mean Attendance']],on='Module Code',how='left')
newdf = newdf.rename(columns={'Mean Attendance': 'Mean Attendance for module'})
newdf.head()
````

````python
plt.scatter(y= newdf['Student overall attendance'], x= newdf['Mean Attendance for module'])
plt.xlabel('avg attendance for module')
plt.ylabel('Student overall attendance')
plt.title('Individual Attendance vs Average Attendance for module ')
````

````python
#pearson correlation
from scipy.stats import pearsonr
corr, p_value = pearsonr(newdf['Mean Attendance for module'],newdf['Student overall attendance'])
print(corr, p_value)
````

````python
#this didnt work, so checking for nan values in each column
newdf[['Mean Attendance for module','Student overall attendance']].isna().sum()
````

````python
#aha!
#replace nan with zero
newdf['Student overall attendance'] = newdf['Student overall attendance'].fillna(0)
#do correlation again
corr, p_value = pearsonr(newdf['Mean Attendance for module'],newdf['Student overall attendance'])
print(corr, p_value)
````

````python
#weak correlation, i assume that that isnt actually a 0.0 p value, but something really small like 1e-200 due to scipy's floating point precision
#anyways, this is evidence against the null hypothesis; its likely that this correlation is not random
#to test again, and because i as a l6 student have to anyway, i'll do linear regression
from sklearn.linear_model import LinearRegression

x = newdf[['Mean Attendance for module']]
y = newdf['Student overall attendance']
model = LinearRegression()
model.fit(x,y)
#intercept and coefficient
print(model.intercept_,model.coef_[0])
````

````python
# y = 0.746..*x+0.4364...
#nice, now to check SSE
y_pred = model.predict(x)
sse = ((y - y_pred) ** 2).sum()
print(sse)
#see markdown below for brief discussion
````

Both the regression and pearson correlation agree that there is a weak positive correlation; the coefficient (slope of the line) used in regression is positive, and so too is the result of the pearson correlation.
The result of the linear regression show us that while there may be a (weak) correlation, a linear model doesn't describe the relationship very well. This is demonstrated by the high sum of squared errors.
